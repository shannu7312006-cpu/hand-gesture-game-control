```python
from pynput import keyboard

def on_press(key):
    try:
        print(f'{key.char} pressed (Ordinal: {ord(key.char):#04x})')
    except AttributeError:
        print(f'{key} pressed')

def on_release(key):
    if key == keyboard.Key.esc:
        # Stop listener
        return False

# Collect events until released
with keyboard.Listener(
        on_press=on_press,
        on_release=on_release) as listener:
    listener.join()
```
